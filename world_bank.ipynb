## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

## Dataset exploration

```python
import pandas as pd
import json
import numpy as np
```

```python
# explore dataframe
df= pd.read_json('data/world_bank_projects.json')
df.head(2)
```

```python
#examine the keys of the DF to identify how to count projects (likely by name)
df.keys()
```

<div class="span5 alert alert-success">
### 10 countries with most projects

```python
# construct a DF grouped by the country name and count the respective projects in each country
project_cnt= df.groupby('countryname')['project_name'].agg('count')

# nlargest applies an order function on the groupby aggregated Series project_cnt
print(project_cnt.nlargest(10))

# the 10 listed results are the 10 countries with the most projects in desceding order
```

<div class="span5 alert alert-success">
### Top 10 major project themes (using column 'mjtheme_namecode')

```python
# explore the mjtheme_namecode of the DF
print(type(df['mjtheme_namecode']))
list_values=[]
list_keys=[]
for i,x in enumerate(df['mjtheme_namecode']):
    for j in range(len(x)):
        list_values.append(x[j]['code'])
        list_keys.append(x[j]['name'])
print(len(list_values), list_values[:10])
print(len(list_keys), list_keys[:10])

#convert list_values form str to int
list_values= [int(i) for i in list_values]
print(type(list_values[0]))
```

```python
# construct a new DF from the lists extracted in previous cell
# new DF is of the theme code and corresponding theme name
theme_df=pd.DataFrame({'code': list_values, 'name':list_keys})
print(theme_df.head())
```

```python
# get the count of each code appearance
cnt_theme_df= theme_df.groupby(['code']).agg('count')

# get the top 10 major themes 
top_ten_mjtheme_df= cnt_theme_df.sort_values('name', ascending=False)
print(top_ten_mjtheme_df.head(10))
```

<div class="span5 alert alert-success">
### Create a dataframe with the missing names filled in

```python
# now we sort the data frame to fill in missing name values
# sort by code 
# AND sort by name -> to get all missing names grouped together, so that they can be reliably and consistently filled
srt_theme_df= theme_df.sort_values(['code', 'name'])
print(srt_theme_df.head())

# realize missing value is actually an empty string '' -> use .replace instead of .fillna
print(srt_theme_df['name'].values)
```

```python
# now we can reliably and consistently fill the empty strings useing .replace and bfill
filled_theme_df= srt_theme_df.replace(to_replace='',method='bfill')
print(filled_theme_df.head())
```
